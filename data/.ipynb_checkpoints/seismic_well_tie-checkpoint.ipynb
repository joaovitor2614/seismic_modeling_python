```python

import os
import numpy as np
import pandas as pd
import welly
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

```

```python
# create an empty list to hold the well objects
wells = []

# read LAS files and create Well objects for each file
las_files = ["3-BRSA-177-ESS_Default_final.las"]
for las_file in las_files:
    well = welly.Well.from_las(las_file)
    wells.append(well)
    

#create a Project object with the list of Well objects
project = welly.Project(wells)

# Set zone interval to get data from wells
basis_depth = range(2400, 2700)
```

```python
desired_curves = ['MD', 'VP', 'RHOB']

BRSA_177_df = project[0].df(keys=desired_curves, basis=basis_depth)

BRSA_177_df
```

# Checar valores brancos para velocidade compressional e densidade

```python
print('Has compressional velocity any missing data', BRSA_177_df['VP'].isna().any())
print('Has density any missing data', BRSA_177_df['RHOB'].isna().any())
```

# Calcular impedância acústica

```python
# Retrieve well logs data as np.arrays from pandas dataframe
compressional_velocity = BRSA_177_df['VP'].values
density = BRSA_177_df['RHOB'].values

acoustic_impedance = compressional_velocity * density
```

# Calculating the reflection coefficients

## The reflexion coefficients are an adimensional parameter. Because of this it's not necessary to be worried about units. The coefficients can be found by the following formula:



```python
def calculate_normal_incidence_reflection_coefs(I):
    reflections_coefficients = []
    for i in range(I.size-1):
        coef=(I[i+1]-I[i])/(I[i+1]+I[i])
        reflections_coefficients = np.append(reflections_coefficients, coef)
        
    return reflections_coefficients

reflections_coefficients = calculate_normal_incidence_reflection_coefs(acoustic_impedance)
```

# Creating a synthetic seismogram with Ricker Wavelet

```python
f = 25
length= 0.299
dt=0.001

def syn_seis(R_coef, f, length, dt):
    
    t = np.arange(-length/2, length/2, dt)
    wavelet = (1.-2.*(np.pi**2)*(f**2)*(t**2))*np.exp(-(np.pi**2)*(f**2)*(t**2))
    
    seismogram = np.convolve(R_coef, wavelet, mode='same')
    
    return seismogram

seismogram = syn_seis(reflections_coefficients, f, length, dt)
seismogram.shape
```

```python
def combo_plot1(top_depth,bottom_depth, RHOB, DEPT, VP, R_coef, seismogram, title):
    
    #plt.hold(True)
    
    fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(16,9), sharey=True)
    fig.suptitle(title, fontsize=22)
    fig.subplots_adjust(top=0.8,wspace=0.1)
    
#General setting for all axis
    for axes in ax:
        axes.set_ylim (top_depth,bottom_depth)
        axes.invert_yaxis()
        axes.yaxis.grid(True)
        axes.get_xaxis().set_visible(False) 
    
#1st track: DT
    
    ax01=ax[0].twiny()
    ax01.set_xlim(1500.0,4500.0)
    ax01.spines['top'].set_position(('outward',0))
    ax01.set_xlabel("Vp [m/s]")
    ax01.plot(VP, DEPT, label='VP[m/s]', color='black')
    ax01.set_xlabel('Vp[m/s]',color='black')    
    ax01.tick_params(axis='x', colors='black')
    ax01.grid(True)
    
#2nd track: RHOB

    ax11=ax[1].twiny()
    ax11.set_xlim(0,3.0)
    ax11.grid(True)
    ax11.spines['top'].set_position(('outward',0))
    ax11.set_xlabel('RHOB[g/cc]', color='red')
    ax11.plot(RHOB, DEPT, label='RHOB[g/cc]', color='red')
    ax11.tick_params(axis='x', colors='red') 
    
#3nd track: COEF
    
    ax21=ax[2].twiny()
    ax21.set_xlim(-0.5,0.5)
    ax21.spines['top'].set_position(('outward',0))
    ax21.set_xlabel("Reflection Coefficients")
    
    x = R_coef
    y = DEPT[1:]
    
    ax21.hlines(y, 0, x, color='blue')
    #ax21.plot(x, y, 'D')  # Stem ends
    ax21.grid(True)   
    
#4nd track: Synthetic Seismogram
    
    ax31=ax[3].twiny()
    ax31.set_xlim(-1,1)
    ax31.spines['top'].set_position(('outward',0))
    ax31.set_xlabel("Synthetic Seismogram")
    ax31.plot(seismogram, DEPT[1:], lw=1, color='black', alpha=0.2)  
    ax31.fill_betweenx(DEPT[1:], seismogram, 0., seismogram > 0., interpolate=False, color='blue', alpha = 0.5)
    ax31.fill_betweenx(DEPT[1:], seismogram, 0., seismogram < 0., interpolate=False, color='red', alpha = 0.5)
    
    ax31.grid(True)  
    
    ax32=ax[3].twiny()
    ax32.set_xlim(-0.8,1.2)
    ax32.xaxis.set_visible(False)
    ax32.plot(seismogram, DEPT[1:], lw=1, color='black', alpha=0.0)  
    ax32.fill_betweenx(DEPT[1:], seismogram, 0., seismogram > 0., interpolate=False, color='blue', alpha = 0.5)
    ax32.fill_betweenx(DEPT[1:], seismogram, 0., seismogram < 0., interpolate=False, color='red', alpha = 0.5)
    
    ax33=ax[3].twiny()
    ax33.set_xlim(-0.6,1.4)
    ax33.xaxis.set_visible(False)
    ax33.plot(seismogram, DEPT[1:], lw=1, color='black', alpha=0.0)  
    ax33.fill_betweenx(DEPT[1:], seismogram, 0., seismogram > 0., interpolate=False, color='blue', alpha = 0.5)
    ax33.fill_betweenx(DEPT[1:], seismogram, 0., seismogram < 0., interpolate=False, color='red', alpha = 0.5)
    
    ax34=ax[3].twiny()
    ax34.set_xlim(-1.2,0.8)
    ax34.xaxis.set_visible(False)
    ax34.plot(seismogram, DEPT[1:], lw=1, color='black', alpha=0.0)  
    ax34.fill_betweenx(DEPT[1:], seismogram, 0., seismogram > 0., interpolate=False, color='blue', alpha = 0.5)
    ax34.fill_betweenx(DEPT[1:], seismogram, 0., seismogram < 0., interpolate=False, color='red', alpha = 0.5)
    
    ax35=ax[3].twiny()
    ax35.set_xlim(-1.4,0.6)
    ax35.xaxis.set_visible(False)
    ax35.plot(seismogram, DEPT[1:], lw=1, color='black', alpha=0.0)  
    ax35.fill_betweenx(DEPT[1:], seismogram, 0., seismogram > 0., interpolate=False, color='blue', alpha = 0.5)
    ax35.fill_betweenx(DEPT[1:], seismogram, 0., seismogram < 0., interpolate=False, color='red', alpha = 0.5)
```

```python
MD = BRSA_177_df.index.to_numpy()
title = "Seismic-well tie for the well Walakpa 1"
bottom_depth = max(MD)
top_depth = min(MD)
combo_plot1(top_depth, bottom_depth, density, MD, compressional_velocity, reflections_coefficients, seismogram, title)
```
