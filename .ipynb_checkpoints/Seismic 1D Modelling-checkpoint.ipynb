# 1D Seismic Modeling Using Well Logs

1D seismic modeling using well logs is a technique in geophysics and petroleum engineering where seismic data is simulated based on information obtained from well logs. Well logs provide detailed information about the subsurface, including rock properties such as velocity, density, and impedance. These properties are crucial for understanding how seismic waves propagate through the Earth's layers.

## Importance

The importance of 1D seismic modeling using well logs lies in its ability to provide insights into subsurface structures and properties without the need for expensive seismic acquisition. Here are some key points about its significance:

### Reservoir Characterization

Well logs offer direct measurements of rock properties at specific depths, allowing for accurate characterization of subsurface reservoirs. By integrating this information into 1D seismic models, engineers and geophysicists can better understand the distribution of hydrocarbons, porosity, and lithology within the reservoir.

### Calibration

Well logs serve as ground truth data for seismic interpretations. By calibrating seismic models with well log data, researchers can ensure that seismic images accurately represent subsurface features. This calibration process enhances the reliability of seismic interpretations and reduces uncertainties in reservoir evaluations.

### Prediction of Seismic Responses

1D seismic modeling enables the prediction of seismic responses at different frequencies and offsets. By incorporating well log data, such as velocity and impedance profiles, into the modeling process, researchers can simulate how seismic waves will behave as they propagate through various geological formations. This predictive capability aids in planning seismic surveys and optimizing data acquisition parameters.

### Exploration and Production Planning

Understanding subsurface geology and fluid properties is crucial for successful exploration and production activities in the oil and gas

```python
def calculate_normal_incidence_reflection_coefs(I):
    reflections_coefficients = []
    for i in range(I.size-1):
        coef=(I[i+1]-I[i])/(I[i+1]+I[i])
        reflections_coefficients = np.append(reflections_coefficients, coef)
        
    return reflections_coefficients

reflections_coefficients = calculate_normal_incidence_reflection_coefs(acoustic_impedance)
```

# Creating Ricker Wavelet 
A wavelet refers to a short, oscillating waveform that is used to represent seismic signals or seismic data. Wavelets are characterized by their duration, frequency content, and shape, and they play a crucial role in various aspects of seismic modeling and seismic data processing.
 
A Ricker wavelet is a signal commonly used in seismic interpretation, modelling and inversion. It has a relatively simple visual form with a positive peak and two negative side-lobes.


```python
peak_dominant_frequency = 30 # in Hz
timeInterval = np.arange(-0.05,0.05, 0.001)

def create_ricker_wavelet(f, t):
    ricker = (1.-2.*(np.pi**2)*(f**2)*(t**2))*np.exp(-(np.pi**2)*(f**2)*(t**2))
    return ricker

ricker_wavelet = create_ricker_wavelet(peak_dominant_frequency, timeInterval)

# Plot wavelet


plt.plot(timeInterval, ricker_wavelet)
plt.xlabel('Time, s')
plt.ylabel('Amplitude')
plt.show()
```

# Seismic Modeling Using the Convolutional Model

Seismic modeling using the convolutional model is a fundamental technique in geophysics and seismic exploration. It involves simulating seismic traces by convolving a wavelet with reflection coefficients that describe the subsurface properties. Here's a detailed overview of this process:

## 1. Wavelet Selection:

- The process begins with selecting an appropriate wavelet. A wavelet is a short, oscillating waveform that represents the seismic source signal or the response of the seismic source to subsurface structures.
- The choice of wavelet depends on factors such as the desired frequency content, phase characteristics, and the geological properties of the subsurface.

## 2. Reflection Coefficients:

- Reflection coefficients describe the amplitude ratio of the reflected wave to the incident wave at each interface between subsurface layers.
- These coefficients are typically calculated based on the acoustic impedance contrast between adjacent layers. The acoustic impedance is the product of the density and seismic velocity of the rock layers.

## 3. Convolution:

- The seismic trace is synthesized by convolving the selected wavelet with the reflection coefficients. Mathematically, convolution is the process of combining two functions to produce a third function that represents how one function modifies the other.
- In the context of seismic modeling, the convolution operation represents the interaction between the wavelet (the source signal) and the reflection coefficients (the subsurface structure).
- At each time sample or depth interval along the seismic trace, the wavelet is convolved with the reflection coefficients to calculate the amplitude of the reflected wave.

## 4. Synthetic Seismic Trace Generation:

- The convolution process is repeated for each time sample or depth interval along the seismic trace, resulting in a synthetic seismic trace that represents the response of the subsurface to the seismic source.
- The synthetic seismic trace contains amplitude variations corresponding to the reflections from subsurface interfaces, allowing geophysicists to interpret the geological features and properties of the subsurface.

## 5. Applications:

- Synthetic seismic traces generated using the convolutional model are valuable for various applications in seismic exploration and reservoir characterization.
- They are used for seismic interpretation, imaging subsurface structures, identifying potential hydrocarbon reservoirs, and optimizing exploration and production strategies.

In summary, seismic modeling using the convolutional model involves synthesizing seismic traces by convolving a wavelet with reflection coefficients that describe the subsurface properties. This technique provides valuable insights into the geological structure and properties of the subsurface, aiding in the exploration and development of hydrocarbon reservoirs.
