1D seismic modeling using well logs is a technique in geophysics and petroleum engineering where seismic data is simulated based on information obtained from well logs. Well logs provide detailed information about the subsurface, including rock properties such as velocity, density, and impedance. These properties are crucial for understanding how seismic waves propagate through the Earth's layers.

## Importance

The importance of 1D seismic modeling using well logs lies in its ability to provide insights into subsurface structures and properties without the need for expensive seismic acquisition. Here are some key points about its significance:

### Reservoir Characterization

Well logs offer direct measurements of rock properties at specific depths, allowing for accurate characterization of subsurface reservoirs. By integrating this information into 1D seismic models, engineers and geophysicists can better understand the distribution of hydrocarbons, porosity, and lithology within the reservoir.

### Calibration

Well logs serve as ground truth data for seismic interpretations. By calibrating seismic models with well log data, researchers can ensure that seismic images accurately represent subsurface features. This calibration process enhances the reliability of seismic interpretations and reduces uncertainties in reservoir evaluations.

### Prediction of Seismic Responses

1D seismic modeling enables the prediction of seismic responses at different frequencies and offsets. By incorporating well log data, such as velocity and impedance profiles, into the modeling process, researchers can simulate how seismic waves will behave as they propagate through various geological formations. This predictive capability aids in planning seismic surveys and optimizing data acquisition parameters.

### Exploration and Production Planning

Understanding subsurface geology and fluid properties is crucial for successful exploration and production activities in the oil and gas

# Calculating the reflections coefficients

Reflection coefficients describe the amplitude ratio of the reflected wave to the incident wave at an interface between two different geological layers. These coefficients play a fundamental role in understanding how seismic waves interact with subsurface structures and are crucial for seismic interpretation and imaging.

Reflection coefficients (ùëÖ) are typically expressed as a dimensionless quantity and are calculated using the acoustic impedance (ùëç) of the two adjacent layers. In a normal incidence context (where the seismic wave strikes the interface perpendicularly), the reflection coefficient for an interface between two layers can be calculated using the following formula:


Where:

ùëÖ is the reflection coefficient.
ùëç1 is the acoustic impedance of the upper layer.
ùëç2 is the acoustic impedance of the lower layer.

```python
def calculate_normal_incidence_reflection_coefs(I):
    reflections_coefficients = []
    for i in range(I.size-1):
        coef=(I[i+1]-I[i])/(I[i+1]+I[i])
        reflections_coefficients = np.append(reflections_coefficients, coef)
        
    return reflections_coefficients

reflections_coefficients = calculate_normal_incidence_reflection_coefs(acoustic_impedance)
```

# Creating Ricker Wavelet 
A wavelet refers to a short, oscillating waveform that is used to represent seismic signals or seismic data. Wavelets are characterized by their duration, frequency content, and shape, and they play a crucial role in various aspects of seismic modeling and seismic data processing.
 
A Ricker wavelet is a signal commonly used in seismic interpretation, modelling and inversion. It has a relatively simple visual form with a positive peak and two negative side-lobes.


```python
peak_dominant_frequency = 30 # in Hz
timeInterval = np.arange(-0.05,0.05, 0.001)

def create_ricker_wavelet(f, t):
    ricker = (1.-2.*(np.pi**2)*(f**2)*(t**2))*np.exp(-(np.pi**2)*(f**2)*(t**2))
    return ricker

ricker_wavelet = create_ricker_wavelet(peak_dominant_frequency, timeInterval)

# Plot wavelet


plt.plot(timeInterval, ricker_wavelet)
plt.xlabel('Time, s')
plt.ylabel('Amplitude')
plt.show()
```
